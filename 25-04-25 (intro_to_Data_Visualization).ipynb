# <b><u>Problem 2 </u></b>

### Read face cream and facewash product sales data and show it using the bar chart. The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

### Expected output:


```python
# Your code here

months = np.arange(1, 13)  
facecream_sales = [2500, 2630, 2140, 3400, 3600, 2760, 2980, 3700, 3540, 1990, 2340, 2900]
facewash_sales = [1500, 1200, 1340, 1130, 1750, 1555, 1120, 1400, 1780, 1890, 2100, 1760]
bar_width = 0.4 

bar1 = months - bar_width / 2
bar2 = months + bar_width / 2

plt.bar(bar1, facewash_sales, width=bar_width, label='facewash', color='dodgerblue')
plt.bar(bar2, facecream_sales, width=bar_width, label='facecream', color='darkorange')

plt.xlabel('Month Number')
plt.ylabel('Product Sales')
plt.title('Face Cream and Facewash Product Sales Month Wise')
 
plt.legend()

plt.show()
```

# <b><u>Problem 3</u></b>

### Display the number of units sold per month for each product using multiline plots. (i.e. separate plotline for each product ).

### Expected Output:



```python

data = df[['month_number', 'facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']]

grouped_data = data.groupby('month_number').sum()

plt.plot(grouped_data.index, grouped_data['facecream'], label='Facecream', color='blue', linestyle='-', marker='o', linewidth=2)
plt.plot(grouped_data.index, grouped_data['facewash'], label='Facewash', color='orange', linestyle='--', marker='o', linewidth=2)
plt.plot(grouped_data.index, grouped_data['toothpaste'], label='Toothpaste', color='green', linestyle=':', marker='o', linewidth=2)
plt.plot(grouped_data.index, grouped_data['bathingsoap'], label='Bathing Soap', color='red', linestyle='-.', marker='o', linewidth=2)
plt.plot(grouped_data.index, grouped_data['shampoo'], label='Shampoo', color='purple', linestyle='-', marker='o', linewidth=2)
plt.plot(grouped_data.index, grouped_data['moisturizer'], label='Moisturizer', color='brown', linestyle='--', marker='o', linewidth=2)

plt.xlabel("Month Number")
plt.ylabel("Total Units Sold")
plt.title("Monthly Sales for Products")

plt.legend()

plt.show()
```
